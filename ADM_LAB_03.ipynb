```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
display(df.head())
```

```python
if 'Cabin' in df.columns:
    df['Cabin'] = df['Cabin'].astype(str)
    le = LabelEncoder()
    df['Cabin_Label'] = le.fit_transform(df['Cabin'])
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
agg_data = df.groupby('Pclass').agg({'Age': 'mean', 'Fare': 'sum'})
```

```python
pivot_table = df.pivot_table(values='Survived', index='Pclass', columns='Sex_male', aggfunc='mean')
```

```python
print("Aggregated Data:")
display(agg_data)
print("Pivot Table:")
display(pivot_table)
```

```python
df.to_csv('/content/titanic.csv', index=False)
```
